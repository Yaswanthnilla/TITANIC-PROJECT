```python
one_hot_encoded_data=pd.get_dummies(df,columns=['embarked','who'])
df1=one_hot_encoded_data
```

```python
temp=df1.drop(['embark_town'],axis=1)
```

```python
temp.head()
```

```python
temp.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X=temp.drop(['survived'],axis=1)
y=temp['survived']
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=44)
```

```python
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
X_train.head()
```

```python
X_train=pd.DataFrame(scale.fit_transform(X_train),columns=X_train.columns)
```

```python
X_train.head()
```

```python
X_test=pd.DataFrame(scale.transform(X_test),columns=X_test.columns)
```

```python
X_test.head()
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
Y_predict=log_model.predict(X_test)
```

```python
acc=log_model.score(X_test,y_test)
acc
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
Y_predict2=knn.predict(X_test)
```

```python
acc2=knn.score(X_test,y_test)
acc2
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
Y_predict3=dtree.predict(X_test)
```

```python
acc3=dtree.score(X_test,y_test)
acc3
```

```python
models = pd.DataFrame({
    'Model': ['KNN', 'Logistic Regression',  
              'Decision Tree'],
    'Score': [acc2, acc, acc3]})
models.sort_values(by='Score', ascending=False)
```
